Read the mushroom dataset

```python
import pandas as pd

df = pd.read_csv("mushrooms.csv")
df
```

Prepare your data. See chap4 for PyML.
    a. Encode categories to numeric form. 

```python
import numpy as np


df = df.drop('stalk-root', 1)
```

```python
class_mapping = {label:idx for idx,label in enumerate(np.unique(df['class']))}
class_mapping
```

```python
df['class'] = df['class'].map(class_mapping)
```

```python

df = pd.get_dummies(df)
df
```

Use model_selection.train_test_split to split your data withholding 20% for testing.

```python
from sklearn.model_selection import train_test_split

trainData, testData = train_test_split(df.values, test_size = 0.20, random_state = 1)
trainData
```

Train and assess/score the accuracy of a decision tree classifier. Comment on your results

```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz

outcomeTrain = [mush[0] for mush in trainData]
samples = [mush[1:] for mush in trainData]



classifier = DecisionTreeClassifier()
classifier = classifier.fit(samples, outcomeTrain)
```

```python
testing = [mushy[1:] for mushy in testData]
outs = [musho[0] for musho in testData]
classifier.score(testing, outs)
```

Prepare data of use with ANN. see chap4 for PyML.

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
sc.fit(samples)
X_train_std = sc.transform(samples)
X_test_std = sc.transform(testing)
```

Train and assess the accuracy of a neural network classifier (multi-layer perceptron). Comment on your results.

```python
from sklearn.linear_model import Perceptron

ppn = Perceptron(n_iter=40, eta0=0.1, random_state=0)
ppn.fit(X_train_std, outcomeTrain)
```

```python
y_pred = ppn.predict(X_test_std)
print('Misclassified samples: %d' % (outs != y_pred).sum())
```

```python
from sklearn.metrics import accuracy_score

print('Accuracy: %.2f' % accuracy_score(outs, y_pred))
```

Use Random forest classifier to select the 5 most important mushroom features

```python
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt

feat_labels = df.columns[1:]

forest = RandomForestClassifier(n_estimators=10000,
                                random_state=0,
                                n_jobs=-1)

from sklearn.model_selection import train_test_split

X_train, X_test = train_test_split(df.values, test_size = 0.20, random_state = 1)

y_train = np.array([mush[0] for mush in X_train])
X_train = np.array([mush[1:] for mush in X_train])


forest.fit(X_train, y_train)
importances = forest.feature_importances_

indices = np.argsort(importances)[::-1]

for f in range(X_train.shape[1]):
    print("%2d) %-*s %f" % (f + 1, 30, 
                            feat_labels[indices[f]], 
                            importances[indices[f]]))

plt.title('Feature Importances')
plt.bar(range(X_train.shape[1]), 
        importances[indices],
        color='lightblue', 
        align='center')

plt.xticks(range(X_train.shape[1]), 
           feat_labels[indices], rotation=90)
plt.xlim([-1, X_train.shape[1]])
plt.tight_layout()
#plt.savefig('./random_forest.png', dpi=300)
plt.show()
```

Discuss how the number of important features impacts the accuracy of the model.

The number of important features is a key factor in determining the accuracy of a model. If you limit the number of important features too much, the model does not have enough questions to ask to create an accurate prediciton. Despite this, it is advantageous to limit the number of important features being used by a model. A model that tests on 5 features is a lot less cumbersome and memory intensive than a model that tests on 22 features. Most of the features are not very influential in determining the classification of an example and thus can usually be eliminated. However, there is a point where too many features have been removed and the accuracy of the model takes a sharp turn downhill.
